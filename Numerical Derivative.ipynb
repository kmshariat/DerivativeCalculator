Derivates are generally calculated analytically by hand. But for digital machines, it has been always a complex task to solve these analytically. We can easily solve these derivatives numerically. Although we will have some error, but different methods, to solving can significantly reduce these errors. The numerical methods to find the first derivative and second derivative is discussed in this notebook. Three numeric approach to calculate the first derivative was discussed and two numeric approach to calculate the second derivative was discussed. 


### First Derivative

Given a continuous function $f(x)$ the first derivative of the function at a point $x=a$ is defined as (in discrete domain)

$$ \left( \frac{df}{dx} \right)^{frwrd} \approx \frac{f(x+dx)-f(x)}{dx} $$
$$ \left( \frac{df}{dx} \right)^{back} \approx \frac{f(x)-f(x+dx)}{dx} $$
$$ \left( \frac{df}{dx} \right)^{cntr} \approx \frac{f(x+dx)-f(x-dx)}{2dx} $$

### Second Derivative

To find the second derivative of the function we can use Three Points or Five points operator given as

$$ \left(  \frac{d^2f}{dx^2} \right)^{3 pnts} \approx \frac{f(x+dx)-2f(x)+f(x-dx}{dx^2} $$
$$ \left(  \frac{d^2f}{dx^2} \right)^{5 pnts} \approx \frac{-\frac{1}{12} f(x-2dx) + \frac{4}{3} f(x-dx)-\frac{5}{2} f(x)+ \frac{4}{3}f(x+dx) - \frac{1}{12}f(x+2dx)}{dx^2} $$

# Plots

```python
#modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
xmax = 10
nx = 100
dx = xmax/(nx-1)

x = np.linspace(0,xmax,nx)

# ------------------------------------------------------------------

#Enter the function here and 
#run all cells to view the plots of the derivatives

f = (1/np.sqrt(2*np.pi*0.2))*np.exp(-(((x-(xmax/2))**2)/(2*0.2)))

# ------------------------------------------------------------------

plt.plot(x,f)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Input Function')
plt.show()
```

```python
nder_frwrd = np.zeros(nx)
for i in range(1,nx-1):
    nder_frwrd[i] = (f[i+1]-f[i])/dx

nder_frwrd[0]=np.NaN
nder_frwrd[nx-1]=np.NaN
    
plt.plot(x,nder_frwrd)
plt.plot(x,f*0, color='gray')
plt.xlabel('x')
plt.ylabel('$f_{prime}(x)$')
plt.title('First Derivative [Forward Method]')
plt.show()
```

```python
nder_back = np.zeros(nx)
for i in range(1,nx-1):
    nder_back[i] = (f[i]-f[i-1])/dx

nder_back[0]=np.NaN
nder_back[nx-1]=np.NaN
    
plt.plot(x,nder_back)
plt.plot(x,f*0, color='gray')
plt.xlabel('x')
plt.ylabel('$f_{prime}(x)$')
plt.title('First Derivative [Backward Method]')
plt.show()
```

```python
nder_cntr= np.zeros(nx)
for i in range(1,nx-1):
    nder_cntr[i] = (f[i+1]-f[i-1])/2*dx

nder_cntr[0]=np.NaN
nder_cntr[nx-1]=np.NaN
    
plt.plot(x,nder_cntr)
plt.plot(x,f*0, color='gray')
plt.xlabel('x')
plt.ylabel('$f_{prime}(x)$')
plt.title('First Derivative [Center Method]')
plt.show()
```

```python
nder_3=np.zeros(nx)      
for i in range (1, nx-1):
    nder_3[i]=(f[i+1] - 2*f[i] + f[i-1])/(dx**2)

nder_3[0]=np.NaN    
nder_3[nx-1]=np.NaN

plt.plot(x,nder_3)
plt.plot(x,f*0, color='gray')
plt.xlabel('x')
plt.ylabel('$f_{dprime}(x)$')
plt.title('Second Derivative [3 Point Method]')
plt.show()
```

```python
nder_5 = np.zeros(nx)      
for i in range (2, nx-2):
    nder_5[i] = (-1/12 * f[i - 2] + 4/3  * f[i - 1] - 5/2 * f[i]  +4/3  * f[i + 1] - 1/12 * f[i + 2]) / dx ** 2
    
nder_5[0]=np.NaN
nder_5[1]=np.NaN
nder_5[nx-1]=np.NaN
nder_5[nx-2]=np.NaN

plt.plot(x,nder_5)
plt.plot(x,f*0, color='gray')
plt.xlabel('x')
plt.ylabel('$f_{dprime}(x)$')
plt.title('Second Derivative [5 Point Method]')
plt.show()
```
